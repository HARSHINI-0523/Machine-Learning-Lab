5. Python program to display the given integer in reverse manner

```python
num=int(input("Enter an Integer: "))
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num//=10
print(f"Reversed Integer: {rev}")
```

```python
num=int(input("Enter an Integer: "))
sum=0
while num>0:
  digit=num%10
  sum+=digit
  num//=10
print(f"Sum of the digits: {sum}")
```

8. Python program to implement matrix addition and multiplication

```python
def matrix_addition(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[i])):
            row.append(A[i][j] + B[i][j])
        result.append(row)
    return result

def matrix_multiplication(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

print("Matrix Addition:")
add_result = matrix_addition(A, B)
for row in add_result:
    print(row)

print("Matrix Multiplication:")
mult_result = matrix_multiplication(A, B)
for row in mult_result:
    print(row)
```

10. Program to swap two numbers using functions

```python
def swap(a,b):
  a,b=b,a
  return a,b
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
a,b=swap(a,b)
print(f"After Swapping: a = {a}, b = {b}")
```

11. Program to calculate the GCD using a recursive function

```python
def gcd(a,b):
  if b==0:
    return a
  else:
    return gcd(b,a%b)
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
res=gcd(a,b)
print(f"The gcd of {a} and {b} is: {res}")
```

12. Program to calculate the Fibonacci Series using recursion

```python
def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)
n=int(input("Enter the number of terms in Fibonacci series: "))
print("Fibonacci Series:")
for i in range(n):
  print(fib(i),end=" ")
```

13. Program to find the Armstrong Number

```python
def is_armstrong(num):
  string=str(num)
  n=len(string)
  sum=0
  for digit in string:
    sum+=int(digit)**n
  return sum==num
a=int(input("Enter the number: "))
if is_armstrong(a):
  print(f"{a} is a Armstrong number")
else:
  print(f"{a} is not a Armstrong number")
```
